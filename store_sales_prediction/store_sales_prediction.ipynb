```python
parts = df['date'].str.split('-', n=2, expand=True)

df['years'] = parts[0].astype(int)
df['month'] = parts[1].astype(int)
df['days'] = parts[2].astype(int)

df.head()
```

```python
df.drop(columns=['date'], inplace=True)
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
scaled=ss.fit_transform(df)
```

```python
x = df.drop(columns=['sales'])
y = df['sales']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_train.shape,x_test.shape
```

```python
y_train.shape , y_test.shape
```

```python
df.columns
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
```

```python
le=LinearRegression()
linear=le.fit(x_train,y_train)
y_pred_linear=linear.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error , mean_absolute_error 
import numpy as np
```

```python
mse=mean_squared_error(y_test,y_pred_linear)
mae=mean_absolute_error(y_test,y_pred_linear)
rmse=np.sqrt(mse)

print("mse",mse)
print("mae",mae)
print("rmse",rmse)
```

```python
from sklearn.ensemble import RandomForestRegressor
Re=RandomForestRegressor()
random=Re.fit(x_train,y_train)
y_pred_random=random.predict(x_test)
```

```python
mse=mean_squared_error(y_test,y_pred_random)
mae=mean_absolute_error(y_test,y_pred_random)
rmse=np.sqrt(mse)

print("mse",mse)
print("mae",mae)
print("rmse",rmse)
```

```python
from xgboost import XGBRegressor
xgb = XGBRegressor(
    n_estimators=300,     # number of trees
    learning_rate=0.05,   # step size
    max_depth=6,          # tree depth
    random_state=42
)

xgb.fit(x_train, y_train)
y_pred_xgb=xgb.predict(x_test)
```

```python
mse=mean_squared_error(y_test,y_pred_xgb)
mae=mean_absolute_error(y_test,y_pred_xgb)
rmse=np.sqrt(mse)

print("mse",mse)
print("mae",mae)
print("rmse",rmse)
```
